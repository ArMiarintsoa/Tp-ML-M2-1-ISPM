```python
import kagglehub
import pandas as pd


# Download latest version
path = kagglehub.dataset_download("vjchoudhary7/customer-segmentation-tutorial-in-python")

print("Path to dataset files:", path)

df = pd.read_csv(path + "/Mall_Customers.csv")
df.head()

```

```python
import sklearn
import pandas as pd

ds = sklearn.datasets.fetch_california_housing()
df = pd.DataFrame(ds.data, columns=ds.feature_names)
print("\n==> Recherche des valeurs manquantes")
print(df.isnull().sum())

print("\n==> Les unités et typage")
print(df.dtypes)

print("\n==> Statistiques descriptives")
print(df.describe())
```

# ***Histogrammes***

```python
import matplotlib.pyplot as plt
df['MedInc'].hist(bins=30)
plt.title("Distribution des revenus")
plt.xlabel("MedInc")
plt.ylabel("Nombre de quartiers")
plt.show()
```

# ***Boxplots***

```python
import seaborn as sns
sns.boxplot(x=df['HouseAge'])
plt.title("Boxplot de l'âge des maisons")
plt.show()
```

# ***Scatter-plots***

```python
# Liste de toutes les colonnes sauf 'Target'
df['Target'] = ds.target
features = [col for col in df.columns if col != 'Target']
print(features)
# Boucle sur chaque variable explicative
for feature in features:
    plt.figure(figsize=(6, 4))  # taille du graphique
    sns.scatterplot(x=df[feature], y=df['Target'])
    plt.title(f"Prix des maisons vs {feature}")
    plt.xlabel(feature)
    plt.ylabel("Prix médian des maisons")
    plt.tight_layout()
    plt.show()
```

```python
df['Target'] = ds.target
sns.scatterplot(x=df['MedInc'], y=df['Target'])
plt.title("Prix des maisons vs Revenu médian")
plt.xlabel("Revenu médian")
plt.ylabel("Prix médian des maisons")
plt.show()
```
